```python
path_test = "/content/drive/My Drive/COS711/"
```

```python
arr = []
for i in range(3100):
  arr.append("")
arrC = 0
gofree = 0
```

```python
with open(path_test + "TrainOneHot.csv") as file:
  read = csv.reader(file, delimiter=',')
  count = 0
  for row in read:
    gofree = 0
    if count == 0:
      #print(f'Column names are {", ".join(row)}')
      count += 1
    else:
      #row[x] to access each variable
      #this is where processing of the images will happen, generate a matrix for the colours
      for i in range(arrC):
        if arr[i] == row[0]:
          gofree = 1
      if gofree == 0:  
        arr[arrC] = row[0]
        arrC += 1
        image = Image.open(path_test + "Train_Images/" + row[0] + ".jpg")
        newImage = image.resize((256,256))
        newImage = newImage.save(path_test + "ResizedTrain/" + row[0] + ".jpg");
        count += 1
        print(count)
```
